# Comment Cleanup for app.js, app.css, and index.html
Remove existing comments from the three files and rewrite the comments cleanly.

```python
from pathlib import Path
import re

workspace = Path(r'c:\Users\jvwsc\Downloads\Prober.in')
files = {
    'app.js': workspace / 'app.js',
    'app.css': workspace / 'app.css',
    'index.html': workspace / 'index.html',
}

def read_file(path):
    return path.read_text(encoding='utf-8')

def write_file(path, content):
    path.write_text(content, encoding='utf-8')
```

```python
def strip_js_css_comments(text):
    result = []
    i = 0
    n = len(text)
    in_string = False
    quote = ''
    escaped = False
    in_line = False
    in_block = False

    while i < n:
        c = text[i]
        nxt = text[i + 1] if i + 1 < n else ''

        if in_block:
            if c == '*' and nxt == '/':
                in_block = False
                i += 2
            else:
                i += 1
            continue

        if in_line:
            if c == '\n':
                in_line = False
                result.append(c)
            i += 1
            continue

        if in_string:
            result.append(c)
            if escaped:
                escaped = False
            elif c == '\\':
                escaped = True
            elif c == quote:
                in_string = False
            i += 1
            continue

        if c in ('"', "'", '`'):
            in_string = True
            quote = c
            result.append(c)
            i += 1
            continue

        if c == '/' and nxt == '*':
            in_block = True
            i += 2
            continue

        if c == '/' and nxt == '/':
            in_line = True
            i += 2
            continue

        result.append(c)
        i += 1

    return ''.join(result)

def strip_html_comments(text):
    return re.sub(r'<!--.*?-->', '', text, flags=re.DOTALL)
```

```python
contents = {name: read_file(path) for name, path in files.items()}

original_sizes = {name: len(content) for name, content in contents.items()}
print('Loaded files:')
for name, size in original_sizes.items():
    print(f' - {name}: {size} bytes')
```

```python
cleaned = {}
cleaned['app.js'] = strip_js_css_comments(contents['app.js'])
cleaned['app.css'] = strip_js_css_comments(contents['app.css'])
cleaned['index.html'] = strip_html_comments(contents['index.html'])
```

```python
def rewrite_js_comments(text):
    header = '// Cleaned app.js: comments removed and rewritten below\n'
    return header + text.lstrip('\n')

def rewrite_css_comments(text):
    header = '/* Cleaned app.css: comments removed and rewritten below */\n'
    return header + text.lstrip('\n')

def rewrite_html_comments(text):
    header = '<!-- Cleaned index.html: comments removed and rewritten below -->\n'
    return header + text.lstrip('\n')

cleaned['app.js'] = rewrite_js_comments(cleaned['app.js'])
cleaned['app.css'] = rewrite_css_comments(cleaned['app.css'])
cleaned['index.html'] = rewrite_html_comments(cleaned['index.html'])
```

```python
for name, path in files.items():
    write_file(path, cleaned[name])
    print(f'Wrote cleaned {name} ({len(cleaned[name])} bytes)')
```

```python
for name in files:
    sample = cleaned[name].splitlines()[:10]
    print(f'--- {name} preview ---')
    print('\n'.join(sample))
    print()
```
